```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
dataset = [
    ['roti', 'susu', 'telur'],
    ['roti', 'mentega', 'roti', 'telur'],
    ['susu', 'mentega', 'roti', 'telur'],
    ['roti', 'susu'],
    ['roti', 'mentega', 'telur']
]
```

```python
te = TransactionEncoder()
te_data = te.fit(dataset).transform(dataset)        #te.fit(dataset) : belajar ada barang apa saja di seluruh transaksi, lalu mentrasform list nya menjadi true/false
df = pd.DataFrame(te_data, columns=te.columns_)     #pd.DataFrame(...): Membungkus hasil matriks tadi ke dalam tabel
df
```

```python
frequent_items = apriori(df, min_support=0.6, use_colnames=True)
frequent_items
```

```python
rules = association_rules(frequent_items, metric='confidence', min_threshold=0.5)
rules
```

```python
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

```python
import matplotlib.pyplot as plt

plt.scatter(rules['support'], rules['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```
