```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/Kuliah/Data-mining/diabetes.csv')

df.head()
df.info()
df.describe()
```

```python
import numpy as np

cols = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']

for col in cols:
    df[col] = df[col].replace(0, np.nan)

df.fillna(df.median(), inplace=True)
df.describe()
# df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Outcome', data=df)
plt.title("Distribusi Pasien Diabetes")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Outcome', data=df)
plt.title("Distribusi Pasien Diabetes")
plt.xlabel("Outcome (0 = Tidak Diabetes, 1 = Diabetes)")
plt.ylabel("Jumlah Pasien")
plt.show()
```

```python
sns.boxplot(x='Outcome', y='Glucose', data=df)
plt.title("Perbandingan Glucose antara Pasien Diabetes dan Non-Diabetes")
plt.show()
```

```python
sns.boxplot(x='Outcome', y='BMI', data=df)
plt.title("BMI terhadap Diabetes")
plt.show()
```
