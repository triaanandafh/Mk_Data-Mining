```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
from google.colab import drive
drive.mount('/content/drive')
# Baca file (sesuaikan nama filenya)
data = pd.read_csv('/content/drive/MyDrive/Kuliah/Data-mining/Advertising.csv')
data.head()
```

```python
data.info()
data.describe()
```

```python
X = data[['Advertising_Budget_USD']]
y = data['Product_Sales_Units']

plt.scatter(X, y)
plt.xlabel('Advertising Budget (USD)')
plt.ylabel('Product Sales (Units)')
plt.title('Advertising Budget vs Product Sales')
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)

print("Model trained successfully!")
print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
```

```python
y_pred = model.predict(X)

plt.scatter(X, y)
plt.plot(X, y_pred)
plt.xlabel('Advertising Budget (USD)')
plt.ylabel('Product Sales (Units)')
plt.title('Linear Regression Result')
plt.show()
```

```python
prediksi = model.predict([[2500]])
print("Prediksi penjualan:", prediksi[0])
```
