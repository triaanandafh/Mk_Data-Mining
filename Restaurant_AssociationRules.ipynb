```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

from google.colab import drive
drive.mount('/content/drive')
# Membaca file CSV
df_raw = pd.read_csv('/content/drive/MyDrive/Kuliah/Data-mining/Restaurant.csv', header=None)

# 2. Memberikan nama kolom
df_raw.columns = ['Borough', 'Cuisine', 'Grade']

# 3. Membersihkan spasi
df_raw = df_raw.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

print(df_raw())
```

```python
# Mengubah kategori menjadi kolom True/False
df_encoded = pd.get_dummies(df_raw)
print(df_encoded.head())
```

```python
# 1. Mencari Itemset yang sering muncul
# Pakai min_support rendah (misal 0.05) karena datanya sangat bervariasi
frequent_itemsets = apriori(df_encoded, min_support=0.05, use_colnames=True)

# 2. Membuat aturan asosiasi
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# 3. Memfilter hasil agar hanya menunjukkan hubungan yang mengarah ke Grade A
hasil_menarik = rules[rules['consequents'].apply(lambda x: 'Grade_A' in str(x))]

print(hasil_menarik[['antecedents', 'consequents', 'support', 'confidence', 'lift']].sort_values(by='confidence', ascending=False))
```
