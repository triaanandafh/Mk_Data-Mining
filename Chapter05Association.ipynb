```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

from google.colab import drive
drive.mount('/content/drive')
# Baca file (sesuaikan nama filenya)
df = pd.read_csv('/content/drive/MyDrive/Kuliah/Data-mining/Chapter05DataSet.csv')

# Kita hanya ambil kolom kategori (Family sampai Support_Group)
# karena kolom Age dan Elapsed_Time tidak cocok untuk Association Rules langsung
df_category = df[['Family', 'Hobbies', 'Social_Club', 'Political', 'Professional', 'Religious', 'Support_Group']]

# Pastikan data berupa 0 dan 1 menjadi True/False
df_encoded = df_category.astype(bool)

print(df_encoded.head())
```

```python
# 1. Cari Itemset
frequent_itemsets = apriori(df_encoded, min_support=0.2, use_colnames=True)

# 2. Buat Rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# 3. Tampilkan hasil
hasil = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
print(hasil.sort_values(by='confidence', ascending=False))
```
