```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
from google.colab import drive
drive.mount('/content/drive')
# Baca file (sesuaikan nama filenya)
data = pd.read_csv('/content/drive/MyDrive/Kuliah/Data-mining/Real estate.csv')
data.head()
```

```python
# eksplorasi
data.info()

# pilih jarak MRT (X) untuk memprediksi harga unit rumah (y)
X = data[['X3 distance to the nearest MRT station']]
y = data['Y house price of unit area']
```

```python
plt.scatter(X, y, color='blue', alpha=0.5)
plt.xlabel("Distance to MRT Station")
plt.ylabel("House Price per Unit Area")
plt.title("MRT Distance vs House Price")
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)

print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
```

```python
y_pred = model.predict(X)

plt.scatter(X, y, color='blue', alpha=0.5)
plt.plot(X, y_pred, color='red', linewidth=2)
plt.xlabel("Distance to MRT")
plt.ylabel("House Price")
plt.title("Linear Regression: Real Estate Prediction")
plt.show()
```

```python
jarak_baru = np.array([[500]])
prediksi = model.predict(jarak_baru)
print(f"Prediksi harga untuk jarak 500m: {prediksi[0]:.2f}")
```

```python
# Baca file dataset insurance
data_ins = pd.read_csv('/content/drive/MyDrive/Kuliah/Data-mining/insurance.csv')
data_ins.head()
```

```python
data_ins.info()
data_ins.describe()
```

```python
# Kita gunakan Age sebagai X dan Charges sebagai y
X = data_ins[['age']]
y = data_ins['charges']
```

```python
plt.scatter(X, y, color='green', alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Medical Charges")
plt.title("Age vs Medical Charges")
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)

print("Intercept:", model.intercept_)
print("Slope (Coefficient):", model.coef_[0])
```

```python
y_pred = model.predict(X)

plt.scatter(X, y, color='green', alpha=0.5)
plt.plot(X, y_pred, color='red', linewidth=2)
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Linear Regression Result: Insurance")
plt.show()
```

```python
usia_baru = np.array([[40]])
prediksi = model.predict(usia_baru)
print(f"Prediksi biaya medis untuk usia 40 tahun: ${prediksi[0]:.2f}")
```
